```python
#  Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#  Step 1: Load the dataset
df = pd.read_csv("C:/Users/ASUS/Documents/CodeVeda/Data Set For Task/1) iris.csv")

#  Step 2: Basic Info
print("Dataset Shape:", df.shape)
print("\nFirst 5 rows:\n", df.head())
print("\nDataset Info:\n")
print(df.info())

#  Step 3: Check Missing Values
print("\nMissing Values:\n", df.isnull().sum())

#  Step 4: Check Duplicates and Remove
duplicates = df.duplicated().sum()
print("\nDuplicate Rows:", duplicates)
df = df.drop_duplicates()

#  Step 5: Summary Statistics
print("\nSummary Statistics:\n", df.describe())

#  Step 6: Correlation Matrix
corr = df.select_dtypes(include=['float64', 'int64']).corr()
print("\nCorrelation Matrix:\n", corr)

#  Step 7: Visualization

# Histogram
df.hist(figsize=(10,6))
plt.suptitle("Feature Distributions", fontsize=14)
plt.show()

# Boxplot
plt.figure(figsize=(10,5))
sns.boxplot(data=df)
plt.title("Boxplot for Outlier Detection")
plt.show()

# Pairplot (scatter + distribution)
sns.pairplot(df, hue=df.columns[-1])  # last column is usually species
plt.show()

# Correlation Heatmap
plt.figure(figsize=(8,5))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
